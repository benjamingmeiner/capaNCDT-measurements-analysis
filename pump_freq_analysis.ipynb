```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import detrend
from utils.plot import plot
from utils.analyze import load_raw_data
%matplotlib notebook
```

```python
directory = "pump_freq"
numbers = range(1,2)
```

```python
data, time, settings = load_raw_data(directory, numbers)
```

```python
data = data[0][0]
```

```python
plt.plot(data)
```

```python
power_spectrum = np.abs(np.fft.rfft(detrend(data)))**2
sampling_time = settings[0]['sampling_time'] / 1000
freqs = np.fft.rfftfreq(data.size, sampling_time)

plt.plot(freqs, power_spectrum)
```
